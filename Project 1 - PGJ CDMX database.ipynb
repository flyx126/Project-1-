```python
#importing our libraries
import pandas as pd
import  matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
#import the files
file="carpetas-de-investigacion-pgj-de-la-ciudad-de-mexico.csv"
categos="categoriasnos.csv"
```

```python
#read the csvs
file_DF=pd.read_csv(file,sep=";")
categos_DF=pd.read_csv(categos, encoding="cp1252")
#verifty the information is ready to work in a dataframe
#file_DF.head()
categos_DF.head()
```

```python
#merge both csv to create one database
mergeDF=file_DF.merge(categos_DF,on="delito",how="inner")
mergeDF
```

```python
#clean the information 
clean_df=pd.DataFrame(mergeDF,columns=[
    "ao_hechos",
    "mes_hechos",
    "delito",
    "categoria_delito",
    "Category",
    "DICCIONARIO",
    "fiscalia",
    "alcaldia_hechos",
    "colonia_hechos"
])

#show the clean DF
clean_df.head()
    
```

```python
#clean row name
clean_df.columns
RenamedDF=clean_df.rename(columns={
    
    'ao_hechos':"Year",
    'mes_hechos': "Month", 
    'delito':"Crime", 
    'categoria_delito':"Legal_Category", 
    'DICCIONARIO':"Our_Category",
    'fiscalia':"Public_Prosecutor", 
    'alcaldia_hechos':"District",
    'colonia_hechos': "Neighborhood"
    
})
RenamedDF

```

```python
#count the total number for crime category
delito=RenamedDF["Legal_Category"].nunique()
delito
```

```python
#count the total number for crime

delitos=RenamedDF["Crime"].nunique()
delitos
```

```python
#show the information of crime to identify the major one in Mexico City
RenamedDF["Crime"].value_counts(ascending=False)
```

```python
#cleaning the information to show only from 2016
filtro=RenamedDF["Year"]>=2016
td_2016=RenamedDF.loc[filtro,:]
td_2016=td_2016[td_2016.District.isin(['TLAHUAC', 'IZTAPALAPA','CUAUHTEMOC','VENUSTIANO CARRANZA',
                                                                'GUSTAVO A MADERO', 'AZCAPOTZALCO','COYOACAN','MIGUEL HIDALGO',
                                                                'TLALPAN','CUAJIMALPA DE MORELOS','XOCHIMILCO','ALVARO OBREGON',
                                                                'BENITO JUAREZ','LA MAGDALENA CONTRERAS', 'MILPA ALTA','IZTACALCO'])]
```

```python
#Year filters
solo_2016=td_2016["Year"]==2016
filtro_2016=td_2016.loc[solo_2016,:]
solo_2017=td_2016["Year"]==2017
filtro_2017=td_2016.loc[solo_2017,:]
solo_2018=td_2016["Year"]==2018
filtro_2018=td_2016.loc[solo_2018,:]
solo_2019=td_2016["Year"]==2019
filtro_2019=td_2016.loc[solo_2019,:]
```

```python
#Crime type filters
data_2016=filtro_2016.groupby("District")
crime_2016=data_2016["Category"].value_counts()
crime_2016= crime_2016.to_frame()
crime_2016=crime_2016.rename(columns={"Category":"Count"})
crime_2016=crime_2016.reset_index()
```

```python
data_2017=filtro_2017.groupby("District")
crime_2017=data_2017["Category"].value_counts()
crime_2017= crime_2017.to_frame()
crime_2017=crime_2017.rename(columns={"Category":"Count"})
crime_2017=crime_2017.reset_index()
```

```python
data_2018=filtro_2018.groupby("District")
crime_2018=data_2018["Category"].value_counts()
crime_2018= crime_2018.to_frame()
crime_2018=crime_2018.rename(columns={"Category":"Count"})
crime_2018=crime_2018.reset_index()
```

```python
data_2019=filtro_2019.groupby("District")
crime_2019=data_2019["Category"].value_counts()
crime_2019= crime_2019.to_frame()
crime_2019=crime_2019.rename(columns={"Category":"Count"})
crime_2019=crime_2019.reset_index()
```

```python
#Plot for type of crime in each District by year
tipo_de_crimen=td_2016["Category"].unique()
for tipo in tipo_de_crimen:
    crime=tipo
    theft_2016=crime_2016["Category"]==crime
    total_theft_2016=crime_2016.loc[theft_2016,:]
    theft_2017=crime_2017["Category"]==crime
    total_theft_2017=crime_2017.loc[theft_2017,:]
    theft_2018=crime_2018["Category"]==crime
    total_theft_2018=crime_2018.loc[theft_2018,:]
    theft_2019=crime_2019["Category"]==crime
    total_theft_2019=crime_2019.loc[theft_2019,:]
    plt.figure(figsize=(20,10))
    ax=plt.subplot()
    p1=plt.bar(total_theft_2016["District"],total_theft_2016["Count"],color="m",width=0.9)
    p2=plt.bar(total_theft_2017["District"],total_theft_2017["Count"],color="b",width=0.7)
    p3=plt.bar(total_theft_2018["District"],total_theft_2018["Count"],color="coral",width=0.5)
    p4=plt.bar(total_theft_2019["District"],total_theft_2019["Count"],color="g",width=0.3)
    plt.xticks(total_theft_2016["District"],rotation=90)
    ax.legend((p1[0],p2[0],p3[0],p4[0]),("2016","2017","2018","2019"))
    ax.set_title(f"Crime type {crime} 2016-2019")
    plt.savefig(f"Plots/Crime Type/{crime}.png",dpi=300)
```

```python
#Plot for each District with total crime count
def ploteRDel(stringDel, newDF, colName):
    minidf=newDF["District"]==stringDel
    minidf=newDF.loc[minidf,:]
    plt.figure(figsize=(10,8))
    ax=plt.subplots()
    plt.bar(minidf[colName],minidf["Count"],color="red")
    plt.xticks(minidf[colName],rotation=90)
    plt.xlim=(minidf["Count"].max())
    plt.xlabel("Type of Crime")
    plt.ylabel("Total Count")
    plt.title(f"Crime Rate: {stringDel} 2019")
    plt.savefig(f"Plots/Delitos_{stringDel}.png",dpi=300)
```
